## define plot map function

```python
def plot_map(data, title=None, europe=True, data_europe=None, vmin=None, vmax=None):
    data, lons_ = add_cyclic_point(data, coord=lons)

    fig = plt.figure(figsize=(12, 6))
    cm = 0.
    ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=cm))

    if europe:
        minlon = -15. + cm
        maxlon = +30. + cm
        ax.set_extent([minlon, maxlon, 35., 70.], ccrs.PlateCarree())

    ax.coastlines()
    ax.set_title(title)

    if europe and data_europe is not None:
        vmin = np.min(data_europe)
        vmax = np.max(data_europe)
    im = ax.pcolormesh(lons_, lats, data, transform=ccrs.PlateCarree(), vmin=vmin, vmax=vmax)
    fig.colorbar(im, ax=ax, pad=0.01)
    plt.show()
```

```python
plot_map(preciparr, europe=True, title='Rainfall Raw Data Example Month', data_europe=data_europe)
```

## Construct time series by loading the whole data

```python
year_range = range(1984,2020)
month_range = range(1,13)
```

```python
data_arrs = []
for y in year_range:
    for m in month_range:
        arr = xr.load_dataset(base_path + str(y) + ("%02d" % m) + ".nc")
        data_arrs.append(arr["precip"].values[0])
data_arrs = np.stack(data_arrs,axis=-1)
data_arrs.shape
```

## substract monthly anomalies

```python
data_anomalies = np.zeros_like(data_arrs)
long_term_means = np.zeros((data_arrs.shape[0],data_arrs.shape[1],12))
for latc in range(data_arrs.shape[0]):
    for lonc in range(data_arrs.shape[1]):
        tseries = data_arrs[latc,lonc].copy()
        tseries_old = tseries.copy()
        for month in range(12):
            monthmask = np.zeros(len(tseries)).astype(bool)
            i = month
            while i < len(tseries):
                monthmask[i] = True
                i += 12
            monthavg = np.average(tseries[monthmask])
            tseries[monthmask] -= monthavg
            long_term_means[latc,lonc,month] = monthavg
        data_anomalies[latc,lonc] = tseries
        """
        plt.figure(figsize=(4,2))
        plt.bar(range(12),long_term_means[latc,lonc])
        plt.show()
        plt.figure(figsize=(12,1.5))
        plt.plot(tseries_old[:50])
        plt.plot(tseries[:50])
        plt.show()
        """
```

## compute all correlations with NAO index
pearson and spearman give very similar results

```python
rs = np.zeros((data_arrs.shape[0],data_arrs.shape[1]))
for latc in range(data_arrs.shape[0]):
    for lonc in range(data_arrs.shape[1]):
        tseries = data_anomalies[latc,lonc]
        #r = stats.pearsonr(tseries,aligned.NAO.values)
        r = stats.spearmanr(tseries, aligned.NAO.values)
        rs[latc,lonc] = r[0]
print("finished")
```

```python
np.max(rs)
```

```python
inds = np.unravel_index(rs.argmax(), rs.shape)
lats[inds[0]].values,lons[inds[1]].values
```

#### sanity check: the highest correlation is obtained at 63.75 north, 338.75 east: right in front of iceland.

```python
plt.figure()
plt.title("Correlations with NAO index")
plt.hist(rs.flatten(),bins=30)
plt.show()
```

## correlations with NAO within europe:

```python
eurrs = -100 * np.ones((data_arrs.shape[0],data_arrs.shape[1]))
for latc in range(data_arrs.shape[0]):
    for lonc in range(data_arrs.shape[1]):
        if lats_mask_europe[latc] and lons_mask_europe[lonc]:
            tseries = data_anomalies[latc,lonc]
            r = stats.spearmanr(tseries, aligned.NAO.values)
            eurrs[latc,lonc] = r[0]
print("finished")
```

```python
plt.figure()
plt.title("Correlations with NAO index in Europe")
plt.hist(eurrs[eurrs > -100].flatten(),bins=30)
plt.show()
```

### seasonal correlations:

```python
#seasons = {'spring' : [3,4,5], 'summer' : [6,7,8], 'autumn' : [9,10,11], 'winter' : [12,1,2]}
seasons = {'winter' : [1,2,12], 'summer' : [7,8]}
```

```python
for sname, smonths in seasons.items():
    print(sname)
    eurrs = -100 * np.ones((data_arrs.shape[0],data_arrs.shape[1]))
    for latc in range(data_arrs.shape[0]):
        for lonc in range(data_arrs.shape[1]):
            if lats_mask_europe[latc] and lons_mask_europe[lonc]:
                tseries = data_anomalies[latc,lonc].copy()
                monthmask = np.zeros(len(tseries)).astype(bool)
                for month in smonths:
                    i = month
                    while i < len(tseries):
                        monthmask[i-1] = True
                        i += 12
                r = stats.spearmanr(tseries[monthmask], aligned.NAO.values[monthmask])
                eurrs[latc,lonc] = r[0]
    print("max correlation: %s" % np.max(eurrs))
    print("finished")
    plot_map(eurrs, title="NAO Europe %s Correlations Map" % sname, \
             vmin = np.min(eurrs[eurrs > -100].flatten()), vmax = np.max(eurrs[eurrs > -100].flatten()))
```
